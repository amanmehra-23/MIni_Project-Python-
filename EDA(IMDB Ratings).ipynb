# Exploratry Data Analysis of Movie Ratings

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv("imdb_top_1000.csv")
```

```python
df = df.drop("Poster_Link", axis=1)
```

```python
df = df.rename(columns={"Series_Title": "Movie_Name"})
```

```python
summary_statistics = df.select_dtypes(include = 'number').dropna().describe()
print(summary_statistics)
```

# Top Movies Based on Different Criteira


```python
highest_rated=df.head(10)[['Movie_Name',"IMDB_Rating"]]
print(highest_rated)
```

```python
plt.bar(highest_rated['Movie_Name'],highest_rated['IMDB_Rating'])
plt.xticks(rotation=45, ha='right')
plt.xlabel('Movie Name')
plt.ylabel('IMDB Rating')
plt.title('Top 10 Movies by IMDB Rating')
plt.show()
```

```python
#df['Gross'] = df['Gross'].str.replace(',', '')
df['Gross'] = df['Gross'].astype('int64')
df['Gross'] = df['Gross'].replace(np.nan, 0)

top_10_movies = df.sort_values(by="Gross", ascending=False).head(10)[['Movie_Name', 'Gross']]
print(top_10_movies)
```

```python
plt.bar(top_10_movies['Movie_Name'],top_10_movies['Gross'],color = "purple")
plt.yscale('log')
plt.yticks([1e6, 2e6, 5e6, 1e7, 2e7, 5e7, 1e8, 2e8, 5e8, 1e9],
           ['1M', '2M', '5M', '10M', '20M', '50M', '100M', '200M', '500M', '1B'])
plt.xticks(rotation=45, ha='right')
plt.xlabel('Movie Name')
plt.ylabel('Gross Revenue')
plt.title('Top 10 Movies by Gross Revenue')
plt.show()
```

```python
plt.boxplot(df["Gross"].dropna())
plt.yscale('log')
plt.yticks([1e6, 2e6, 5e6, 1e7, 2e7, 5e7, 1e8, 2e8, 5e8, 1e9],
           ['1M', '2M', '5M', '10M', '20M', '50M', '100M', '200M', '500M', '1B'])

plt.xlabel('Gross Revenue')
plt.ylabel("Gross Revenue in 10 Millions")
plt.title("Distribution of Gross Earnings of Top 1000 movies on IMDB")
plt.show()
```

```python
most_voted = df.sort_values(['No_of_Votes'], ascending = False).head(10)
plt.bar(most_voted['Movie_Name'],most_voted['No_of_Votes'], color="orange")
plt.xticks(rotation=45, ha='right')
plt.xlabel('Movie Name')
plt.ylabel('Number of Votes')
plt.title('Top 10 Movies by Number of Votes')
plt.show()
```

```python

bins = [6, 7, 8, 9, 10]
labels = ['6-7', '7-8', '8-9', '9-10']
df['RatingRange'] = pd.cut(df['IMDB_Rating'], bins=bins, labels=labels, include_lowest=True)
rating_counts = df['RatingRange'].value_counts().sort_index()


plt.bar(rating_counts.index.astype(str), rating_counts, color="green")


plt.xlabel('Rating Range')
plt.ylabel('Number of Movies')
plt.title('Number of Movies in Each Rating Range')

plt.show()
```

```python
plt.bar(df['Director'].value_counts().head(10).index ,df['Director'].value_counts().head(10), color='pink' )
plt.xticks(rotation=45, ha='right')
plt.xlabel('Director')
plt.ylabel('Number Of Movies')
plt.title('Most Movies in Top 1000 by Directors')
plt.show()
```

```python
top_directors = df['Director'].value_counts().head(10)

plt.bar(top_directors.index, top_directors, color='red')

plt.xticks(rotation=45, ha='right')
plt.xlabel('Director')
plt.ylabel('Number Of Movies')
plt.title('Top 10 Directors with Most Movies')

plt.show()
```

```python

top_rated_movies = df.sort_values(by='IMDB_Rating', ascending=False).head(25)

plt.figure(figsize=(10, 6))
plt.scatter(top_rated_movies["IMDB_Rating"], top_rated_movies["Runtime"], color='skyblue', s=100)

plt.xlabel("IMDb Rating")
plt.ylabel("Runtime (mins)")
plt.title("Top Rated Movies: IMDb Rating vs. Runtime")

plt.show()



```

```python
certificate_counts=df["Certificate"].fillna("Not Rated").value_counts()
plt.figure(figsize=(8,8))
certificate_counts = certificate_counts[certificate_counts / certificate_counts.sum() * 100 >= 2]

plt.pie(certificate_counts, labels=certificate_counts.index, autopct='%1.1f%%', startangle=90, colors=['gold', 'pink', 'lightgreen', 'skyblue'])
plt.title('Distribution of Certificates in Top 1000 Movies')

plt.show()
```

```python

df['Released_Year'] = pd.to_numeric(df['Released_Year'])
df['Decade'] = (df['Released_Year'] // 10) * 10
decade_counts = df['Decade'].value_counts().sort_index()

plt.bar(decade_counts.index, decade_counts, width=8, color='gold')

plt.xticks(rotation=45, ha='right')
plt.xlabel('Decade')
plt.ylabel('Number Of Movies')
plt.title('Best Movies Released Over Each Decade')

plt.show()
```

```python
genres = []
for genre_list in df['Genre'].str.split(','):
    genres.extend([genre.strip().lower() for genre in genre_list])

genre_counts = pd.Series(genres).value_counts().sort_values(ascending=False)
top_genres = genre_counts.head(10)

fig, axs = plt.subplots(figsize=(10, 8))
top_genres.plot(kind='barh', color='teal')
plt.xlabel('Number of Movies')
plt.ylabel('Genres')
plt.title('Number of Movies for Each Genre' )
plt.show()
```
